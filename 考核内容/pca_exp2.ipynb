```python
import pandas as pd #用于数据预处理
import numpy as np #python中用于数据计算的核心库
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import load_breast_cancer #加载一个乳腺癌的数据集
```

```python
cancer=load_breast_cancer() #读取数据
```

```python
cancer.keys()
```

```python
#print(cancer['DESCR']) #展示数据集内容
```

```python
df=pd.DataFrame(cancer['data'],columns=cancer['feature_names'])
#cancer['data']提取了数据集中的特征部分（即数值数据）
#columns=cancer['feature_names']指定了DataFrame的列名，使用数据集中的特征名称
```

```python
df.head()
```

```python
from sklearn.preprocessing import StandardScaler #用于数据缩放
```

```python
scaler=StandardScaler()#创建StandarScaler的实例
scaler.fit(df) #使用fit方法计算DataFrame df中每个特征的均值和标准差
```

```python
scaled_data=scaler.transform(df)
#使用scaler的transform方法，将DataFrame df中的特征数据进行标准化
#每个特征将根据先前计算的均值和标准差进行转换
scaled_data
```

```python
from sklearn.decomposition import PCA
```

```python
pca=PCA(n_components=2) #创建一个PCA实例pca，设置n_components=2，表示希望将数据降维到2个主成分
```

```python
pca.fit(scaled_data)#使用fit方法对标准化后的数据scaled_data进行训练，以计算主成分。
```

```python
x_pca=pca.transform(scaled_data)
#使用pca的transform方法，将标准化后的数据scaled_data映射到新的主成分空间
#结果x_pca将包含每个样本在前两个主成分上的坐标，便于后续的可视化和分析
```

```python
scaled_data.shape #返回一个元组，表示scaled_data的维度
```

```python
plt.figure(figsize=(8,6))#创建一个新的图形，设置图形的大小为8英寸宽和6英寸高
plt.scatter(x_pca[:,0],x_pca[:,1],c=cancer['target'])
#绘制散点图，其中x_pca[:, 0]表示第一个主成分，x_pca[:, 1]表示第二个主成分
#c=cancer['target']用于根据目标变量（即肿瘤类型）为每个点指定颜色
plt.xlabel('First principle component')
plt.ylabel('Second principle component')
```
