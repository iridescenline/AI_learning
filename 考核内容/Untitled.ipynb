```python
from sklearn import datasets
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.neighbors import KNeighborsClassifier  #ç›´æ¥ä½¿ç”¨ KNeighborsClassifierï¼Œå› ä¸ºå®ƒå·²ç»å®ç°äº† KNNï¼Œå¹¶ä¸”æä¾›äº†æ–¹ä¾¿çš„æ¥å£è¿›è¡Œäº¤å‰éªŒè¯
import numpy as np

# åŠ è½½æ•°æ®é›†
iris = datasets.load_iris() 
X = iris.data
y = iris.target

# è®¾ç½® k çš„èŒƒå›´
k_values = range(1, 11) #å®šä¹‰äº†ä¸€ä¸ªä» 1 åˆ° 10 çš„ kå€¼èŒƒå›´
mean_accuracies = []

# å¯¹æ¯ä¸ª k å€¼åˆ›å»ºä¸€ä¸ª KNeighborsClassifier å®ä¾‹ï¼Œè¿›è¡Œ 5 æŠ˜äº¤å‰éªŒè¯ï¼Œå¹¶è®°å½•æ¯ä¸ª ğ‘˜ å€¼ çš„å¹³å‡å‡†ç¡®ç‡ã€‚
for k in k_values:
    knn = KNeighborsClassifier(n_neighbors=k)
    scores = cross_val_score(knn, X, y, cv=5)  # 5æŠ˜äº¤å‰éªŒè¯
    mean_accuracies.append(scores.mean())

# æ‰¾åˆ°æœ€ä½³ k å€¼
best_k = k_values[np.argmax(mean_accuracies)]
best_accuracy = max(mean_accuracies)

print(f"æœ€ä½³ k å€¼: {best_k}ï¼Œå¹³å‡å‡†ç¡®ç‡: {best_accuracy:.3f}")
```
